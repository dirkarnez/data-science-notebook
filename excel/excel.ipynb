```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

excel_data_df = pandas.read_excel('sample.xlsx', sheet_name='GDP')

plt.rcdefaults()
figure, axes  = plt.subplots()

countries = excel_data_df['Country']
y_pos = np.arange(len(countries))

axes.barh(y_pos, excel_data_df['GDP (trillion)']) # make a horizontal bar plot
axes.set_yticks(y_pos) # set the y ticks with list of ticks
axes.set_yticklabels(countries) # set the y-tick labels with list of strings labels.
axes.invert_yaxis()  # invert the y-axis.

axes.set_xlabel('GDP in trillion dollars')
axes.set_title('GDP of countries')

plt.show()
```
